```python
import random
random.shuffle(iris_data)
n = len(iris_data)
train = iris_data[:int((0.7)*n)]
test = iris_data[int((0.7)*n):]
```

```python
#Create a new network
net4=network.Network([4,20,3])
```

```python
results=net4.SGD(train, 50, 5, 0.1,test)
```

```python
train_r=results[0]
test_r=results[1]
```

```python
import matplotlib.pyplot as plt
plt.plot(np.array(train_r)[:,2])
plt.plot(np.array(train_r)[:,3])
plt.plot(np.array(train_r)[:,4])
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('Error by Cost Function (training)')
plt.legend(("MSE","CrossEntropy","LogLikelihood"))
plt.show()
```

```python
plt.plot(np.array(train_r)[:,2])
plt.plot(np.array(test_r)[:,2])
plt.legend(('Train', 'Test'))
plt.ylabel('Error')
plt.xlabel('Epoch')
plt.title('Train vs. Test: MSE')
plt.show()
```

```python
plt.plot(np.array(train_r)[:,3])
plt.plot(np.array(test_r)[:,3])
plt.legend(('Train', 'Test'))
plt.ylabel('Error')
plt.xlabel('Epoch')
plt.title('Train vs. Test: CrossEntropy')
plt.show()
```

```python
plt.plot(np.array(train_r)[:,4])
plt.plot(np.array(test_r)[:,4])
plt.legend(('Train', 'Test'))
plt.ylabel('Error')
plt.xlabel('Epoch')
plt.title('Train vs. Test: LogLikelihood')
plt.show()
```
